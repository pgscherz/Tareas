# Tarea 5.10

Un sensor recibe en promedio $50$ partículas por minuto. Ese sensor está diseñado tal que puede detectar si fue impactado por al menos una partícula, pero no tiene ninguna información sobre cuántas partículas lo impactaron.

El sensor se encendió durante un segundo e indicó impacto. Grafique la distribución de probabilidad de que ese pixel haya sido impactado por $n$ partículas.

## Solución   

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Como la media es de 50 partículas por minuto, supongo que para un segundo la media será de 5/6.
lamda = 5/6

#Para variar la cantidad de partículas las cuales queremos saber su probabilidad de detección.
particulas = 7
```

```python
def poisson(n, lamda):
    '''Esta función entrega la probabilidad de detección de algún número de partículas'''
    # escribimos (n+1) en lugar de n, porque el sensor detecta desde una partícula en adelante.
    return lamda**(n+1)*np.exp(-lamda)/np.math.factorial(n+1)
```

```python
#Calculamos la probabilidad de detección para cada cantidad de partículas.
predicciones = np.array([poisson(n, lamda)for n in range(particulas)])
p= np.linspace(1,particulas,particulas)
```

```python
#Graficamos la distribución. 
plt.bar(p, predicciones)
plt.ylabel('Probabilidad de Detección')
plt.xlabel('Número de partículas')
plt.show()
```
